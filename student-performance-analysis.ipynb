## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import (read) Dataset

Dataset link: https://www.kaggle.com/datasets/devansodariya/student-performance-data/data

**About the dataset**

Student Performance Data was obtained in a survey of students' math course in secondary school.
It consists of 33 Column. Dataset Contains Features like,

- school ID

- gender

- age

- size of family

- Father education

- Mother education

- Occupation of Father and Mother

- Family Relation

- Health

- Grades

### Add another column for final grade

```python
df['G1'] = pd.to_numeric(df['G1'], errors='coerce')
df['G2'] = pd.to_numeric(df['G2'], errors='coerce')
df['G3'] = pd.to_numeric(df['G3'], errors='coerce')

df['Final_Grade'] = df[['G1', 'G2', 'G3']].mean(axis=1)
```

## Relationship of Grades among study time, absences, failures, age, free time, health, travel time, and time to go out

```python
numeric_cols = [
    'G1', 'G2', 'G3', 'Final_Grade',
    'studytime', 'absences', 'failures',
    'age', 'freetime', 'health', 'traveltime', 'goout'
]
corr_matrix = df[numeric_cols].corr(method='pearson')
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

```python
final_grade_corr = corr_matrix['Final_Grade'].sort_values(ascending=False)
final_grade_corr
```

```python
spearman_corr = df[numeric_cols].corr(method='spearman')
spearman_corr['Final_Grade'].sort_values(ascending=False)
```

The correlation analysis reveals that prior academic performance is the strongest predictor of students’ final grades. Both Pearson and Spearman correlations show very strong positive relationships between Final_Grade and earlier assessment scores (G1, G2, and G3), with G2 and G3 exhibiting the highest correlations, indicating that consistent performance throughout the academic year strongly influences final outcomes. Studytime shows a weak but positive association with Final_Grade, suggesting that increased study time contributes modestly to higher achievement. In contrast, the number of past failures demonstrates a moderate negative correlation with Final_Grade, highlighting its detrimental impact on academic performance. Other factors such as absences, age, travel time, going out with friends, health, and free time exhibit weak or negligible correlations, implying limited direct influence on final grades. Overall, the findings suggest that academic history and learning behaviors are more strongly related to student achievement than demographic or lifestyle factors.
